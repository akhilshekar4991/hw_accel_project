```python
# code taken from https://stackoverflow.com/questions/55897076/creating-a-plot-of-frequency-of-items-in-a-list-in-python

from collections import Counter
import pandas
import json
from tqdm import tqdm

input_file = open('build/vanilla_mapping.json')
jsonData = json.load(input_file)

json_vector = jsonData["vector_lengths"]
#for items in tqdm(jsonData):
#    json_vector.append(items)

print(f'Length of JSON vector = {len(json_vector)}')
```

```python
count = Counter(json_vector)
print(type(count))
print(count.values())
df = pandas.DataFrame.from_dict(count, orient='index')
df.plot(kind='bar', xlabel='Bucket Length', ylabel='Frequency')

#print(data)
#type(count)
```
